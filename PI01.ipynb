```python
#importamos las librerías necesarias para realizar nuestro procesamiento de datos
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#Leemos los datasets entregado
dataMovies = pd.read_csv("./Dataset/movies_dataset_v1.csv",  encoding='latin-1')
dataCredits = pd.read_csv("./Dataset/credits.csv", encoding='latin-1')
```

```python
#dataMovies.shape
dataCredits.shape
```

```python
#primero vamos a concatenar los dataframes y luego a realizar las transformaciones
data = pd.concat([dataMovies, dataCredits], axis=1)
```

```python
 # Checkeamos Nulos por columna revenue
data['revenue'].isna().sum() 
```

```python
 # Checkeamos Nulos por columna budget
data['budget'].isna().sum() 
```

```python
#Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.
data['revenue'].fillna(0, inplace=True)
data['budget'].fillna(0, inplace=True)
```

```python
#Chequeo que no hayan quedado nulos
data['budget'].isna().sum() 
```

```python
#Los valores nulos del campo release date deben eliminarse.
 # Checkeamos Nulos por columna release
data['release_date'].isna().sum() 
```

```python
#Elimino los nulos del campo release_date
data.drop(dataMovies[dataMovies['release_date'].isna()].index, inplace=True)
```

```python
data.info()
```

```python
#df.dtypes
#Verificamos los tipos de datos

#df['columna de fecha'] = pd.to_datetime(df['columna de fecha'])   
# Las columnas de fecha suelen venir en type object, es bueno cambiarlas a type datetime
#dataMovies['release_date'].dtype 
data['release_date'] = pd.to_datetime(data['release_date'], format= "%Y-%m-%d", errors='coerce')
data.info()
```

```python
#data['release_date'].isna().sum() 
#Elimino los nuevos nulos generados al convertir el tipo de dato del campo release_date
data.drop(data[data['release_date'].isna()].index, inplace=True)
```

```python
data['release_date'].isna().sum() 
```

```python
#además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

data['release_year'] = data['release_date'].dt.year.astype(int)
```

```python
data.info()
```

```python
data.head(20)
```

```python
#Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, 
#dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.


# Algunas columnas tienen numeros pero son de tipo string(budget), con to_numeric las cambiamos a integer

data['budget'] = pd.to_numeric(data['budget'])
data['return'] =  data.apply(lambda row: row['revenue'] / row['budget'] if (pd.notnull(row['revenue']) and pd.notnull(row['budget'])  and (row['budget']) != 0) else 0, axis=1)
```

```python
data.head(20)
```

```python
#Eliminar las columnas que no serán utilizadas, video,imdb_id,adult,original_title,poster_path y homepage.
data.drop(columns=['video', 'imdb_id', 'adult','original_title','poster_path', 'homepage'], inplace=True)
```

```python
data.columns
```

```python
# Desanidar el campo 'belongs_to_collection'

import ast

data['franquicia'] = data['belongs_to_collection'].apply(lambda x: ast.literal_eval(x)['name'] if isinstance(x, str) else '').tolist()
                                                         
# Imprimir el DataFrame resultante
data.head(5)
```

```python
#desanidar genres
import json

data['genres'] = data['genres'].str.replace("'", '"')
data['genres'] = data['genres'].apply(lambda x: [item['name'] for item in json.loads(x)]).tolist()

data
```

```python
#desanidar production_companies

# convertir strings en 'production_companies' a listas de dictionarios
data['production_companies'] = data['production_companies'].apply(ast.literal_eval)

# Funcion para extraer los nombres de las productoras a una lista
def get_company_names(row):
    if isinstance(row, list):
        return [company['name'] for company in row]
    else:
        return []  # Or handle it as per your requirement

# Crear columna 'productoras'
data['productoras'] = data['production_companies'].apply(get_company_names)

data.head(5)
```

```python
#desanidar production_countries

#convertir strings en 'production_countries' a listas de dictionarios
data['production_countries'] = data['production_countries'].apply(ast.literal_eval)

# Funcion para extraer los nombres de las productoras a una lista
def get_company_names(row):
    if isinstance(row, list):
        return [company['name'] for company in row]
    else:
        return []  # Or handle it as per your requirement

# Crear columna 'paises'
data['paises'] = data['production_countries'].apply(get_company_names)

data.head(5)
```

```python
#desanidar crew (Directores)

data['crew'] = data['crew'].apply(ast.literal_eval)

# Funcion para extraer los nombres de las productoras a una lista
def get_director_names(row):
    if isinstance(row, list):
        return [company['name'] for company in row if company['job'] == 'Director']
    else:
        return []  # Or handle it as per your requirement

# Crear columna 'directores'
data['directores'] = data['crew'].apply(get_director_names)

data.head(5)
```
